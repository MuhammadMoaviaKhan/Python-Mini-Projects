```python
i = 0
life_line_one = 0
life_line_two = 0
total_win = 0
one = 1
onee = 1
print('\n **Who Wants to Be a Millionaire?**')
print('\n')
for que in questions:

  print('Life Lines')
  print(f'Enter 1 for 50 50 Life line remain: {one}')
  print(f'Enter 2 for Audience support Life line remain: {onee}')
  print('\n')
  print(f"Question: {que['question']}")
  print(f"Choices: {que['choices']}")
  print(f"Prize: {que['prize']}")
  answer = input('Enter the answer: ').lower()


  if answer == '1' and life_line_one == 0:
    life_line_one += 1
    one -= 1
    redesign = game_data['questions'][i]['choices']
    for j in game_data['questions'][i]['choices']:
      if j == correct_answers[i]:
        game_data['questions'][i]['choices'].remove(j)

    wrong_1 = random.choice(game_data['questions'][i]['choices'])
    game_data['questions'][i]['choices'].remove(wrong_1)

    wrong_2 = random.choice(game_data['questions'][i]['choices'])
    game_data['questions'][i]['choices'].remove(wrong_2)

    n = random.randint(0, 1)
    game_data['questions'][i]['choices'].insert(n, correct_answers[i])
    game_data['questions'][i]['choices'] = redesign
    print(game_data['questions'][i]['choices'])

    life_line_1 = input('Enter the answer: ').lower()

    if life_line_1 == correct_answers[i].lower():
      total_win +=  que["prize"]
      print(f'This is the correct answer, You won Rs{total_win}.')

    else:
      print(f'This is the wrong answer, You won Rs{total_win}. \n')
      if i == 0 or i == 1:
        print(f'You give {i} answer correctly.')
      else:
        print(f'You give {i} answers correctly.')
        break
  elif answer == '1' and life_line_one == 1:
    print("""You already used your 50 50 Life Line. Don't try to get another time this 50 50 Life Line. This is against the rule and you lose your game """)
    answer = input('Enter the answer: ').lower()

    if answer == correct_answers[i].lower():
      total_win +=  que["prize"]
      print(f'This is the correct answer, You won Rs{total_win}.')
      print('\n\n')
      i += 1
      continue
    else:
      print(f'This is the wrong answer, You won Rs{total_win}. \n')
      if i == 0 or i == 1:
        print(f'You give {i} answer correctly.')
      else:
        print(f'You give {i} answers correctly.')
    break

  elif answer == '2' and life_line_two == 0:
    life_line_two += 1
    onee -= 1
    for k in game_data['questions'][i]['choices']:
      if k == correct_answers[i]:
        game_data['questions'][i]['choices'].remove(k)

    game_data['questions'][i]['choices'].insert(1, correct_answers[i])
    print(game_data['questions'][i]['choices'])
    rndm_1 = random.randint(1, 3)
    rndm_2 = random.randint(8, 12)
    rndm_3 = random.randint(4, 10)
    rndm_4 = random.randint(3, 4)
    kkk = random.randint(23, 30)
    life_line_2 = random.choices(game_data['questions'][i]['choices'], weights = [rndm_1, rndm_2, rndm_3, rndm_4], k = 23)

    for l in enumerate(life_line_2, start = 1):
      print(l)

    answer_in_lifeline = input('Enter the answer: ').lower()

    if answer_in_lifeline == correct_answers[i].lower():
      total_win +=  que["prize"]
      print(f'This is the correct answer, You won Rs{total_win}.')

    else:
      print(f'This is the wrong answer, You won Rs{total_win}. \n')
      if i == 0 or i == 1:
        print(f'You give {i} answer correctly.')
      else:
        print(f'You give {i} answers correctly.')
      break

  elif answer == '2' and life_line_two == 1:
    print("""You already used your Audience support Life Line. Don't try to get another time this Audience support Life Line. This is against the rule and you lose your game """)
    answer = input('Enter the answer: ').lower()

    if answer == correct_answers[i].lower():
      total_win +=  que["prize"]
      print(f'This is the correct answer, You won Rs{total_win}.')
      print('\n\n')
      i += 1
      continue
    else:
      print(f'This is the wrong answer, You won Rs{total_win}. \n')
      if i == 0 or i == 1:
        print(f'You give {i} answer correctly.')
      else:
        print(f'You give {i} answers correctly.')
    break


  elif answer == correct_answers[i].lower():
    total_win +=  que["prize"]
    print(f'This is the correct answer, You won Rs{total_win}.')
  else:
    print(f'This is the wrong answer, You won Rs{total_win}. \n')
    if i == 0 or i == 1:
      print(f'You give {i} answer correctly.')
    else:
      print(f'You give {i} answers correctly.')
    break

  print('\n\n')
  i += 1
```
